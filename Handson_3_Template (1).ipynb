```python
def is_even(n: int) -> bool:
    return n % 2 == 0

def main():
    print("Even or Odd Checker\n")
    num = int(input("Enter an integer: "))
    if is_even(num):
        print("This is an even number.")
    else:
        print("This is an odd number.")

main()
```

```python
import conversions

def display_title():
    print("Feet and Meters Converter\n")

def display_menu():
    print("Conversions Menu:")
    print("a.\tFeet to Meters")
    print("b.\tMeters to Feet")

def main():
    display_title()
    again = "y"
    while again.lower() == "y":
        display_menu()
        choice = input("Select a conversion (a/b): ").strip().lower()
        print()

        if choice == "a":
            feet = float(input("Enter feet: "))
            meters = conversions.feet_to_meters(feet)
            print(f"{meters:.2f} meters")
        elif choice == "b":
            meters = float(input("Enter meters: "))
            feet = conversions.meters_to_feet(meters)
            print(f"{feet:.2f} feet")
        else:
            print("Invalid selection")

        print()
        again = input("Would you like to perform another conversion? (y/n): ").strip().lower()
        print()

    print("Thanks, bye!")

main()
```

```python
import sales_tax as st

def display_title():
    print("Sales Tax Calculator\n")

def get_total():
    print("ENTER ITEMS (ENTER 0 TO END)")
    total = 0.0
    while True:
        cost = float(input("Cost of item: "))
        if cost == 0:
            break
        total += cost
    return total

def main():
    display_title()
    again = "y"
    while again.lower() == "y":
        total = get_total()
        tax = st.calculate_sales_tax(total)
        total_after = st.total_after_tax(total)

        print(f"Total:\t {total:.2f}")
        print(f"Sales tax:\t{tax:.2f}")
        print(f"Total after tax: {total_after:.2f} \n")

        again = input("Again? (y/n): ").strip().lower()
        print()

    print("Thanks, bye!")

main()
```

```python
def get_valid_int() -> int:
    while True:
        n = int(input("Please enter an integer between 1 and 5000: "))
        if 1 <= n <= 5000:
            return n
        print("Invalid integer. Please try again.")

def count_factors(n: int) -> int:
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    return count

def main():
    print("Prime Number Checker\n")
    again = "y"
    while again.lower() == "y":
        n = get_valid_int()
        factors = count_factors(n)

        if factors == 2:
            print(f"{n} is a prime number.\n")
        else:
            print(f"{n} is NOT a prime number.")
            print(f"It has {factors} factors.\n")

        again = input("Try again? (y/n): ").strip().lower()
        print()

    print("Bye!")

main()
```
